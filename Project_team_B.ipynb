```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as plot
df = pd.read_csv('/content/covid19.csv')
df.head()
```

```python
#removing incorrect data (data with no countries)
df2=df.drop([221,222,223,224,225,226,227,228],axis=0)

#replacing null values with 0
df2=df2.fillna(0)
```

```python
#set 0 in all TotalDeaths that cannot be replaced with fillna
for i in range (220):
  if df2['TotalDeaths'][i]==' ':
    df2['TotalDeaths'][i]=0
df2
```

```python
df2['TotalDeaths']=df2['TotalDeaths'].astype('f')
df2.info()
```

```python
df2.plot(kind='line',x='Country,Other',y='TotalDeaths')
plt.show
```
